```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook.**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

* **Create a figure object called `fig` using `plt.figure()`.**
* **Use `add_axes` to add an axis to the figure canvas at [0,0,1,1]. Call this new axis `ax`.**
* **Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel("x")
ax.set_ylabel("z")
ax.set_title("zoom")

ax2.plot(x,y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
sns.set()
```

```python
# code here to load the dataset as a dataframe from seaborn datasets
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.barplot(x='species', y='flipper_length_mm', data=penguins,palette=["blue","orange","green"])
plt.title("Flipper Length for penguin species")
plt.show()
```
